```python
# Normal Sci. Comp. Imports
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
# create random number generator for later/throughout
```

## 30 Epochs Data plots

#### Import the data from the .txt file using numpy.loadtxt first and parse it

```python
# first import the epochs data - Train and Validation accucary are in the 2nd and fourth column
X_eps = np.loadtxt(fname='TF_datasets/FFNN_30epochs_Tr&Val_data.txt')[:, [1, 3]]
```

#### Now plot the data to a figure

```python
Eps_fig = plt.figure()
plt.plot(np.arange(1,31), X_eps[:, 0]*100, 'k.--', np.arange(1,31), X_eps[:, 1]*100, 'b.--', markersize=16)
plt.xlabel('Epochs', fontsize=16)
plt.ylabel('% Accuracy', fontsize=16)
plt.legend(['Train', 'Validation'], fontsize=16)
plt.show()
```

#### Save the above epochs figure

```python
plt.savefig(fname='TF_plots/30epochs.png', format='png')
```

## Variable Num Layers Data

#### Import all the data from the txt file and then part it as needed

```python
# Get all the traing and validation accuracy data
layer_tr = np.loadtxt(fname='TF_datasets/FFNN_VarLays_Acc_data.txt')[:, [0, 2, 4, 6, 8]]
layer_val = np.loadtxt(fname='TF_datasets/FFNN_VarLays_Acc_data.txt')[:, [1, 3, 5, 7, 9]]
```

#### Plot the Layers Data to a fig

```python
layer_fig = plt.figure()
plt.plot(np.arange(2,7), layer_tr[-1, :]*100, 'k.--', np.arange(2, 7), layer_val[-1,:]*100, 'b.--', markersize=16)
plt.xlabel('Num Layers', fontsize=16)
plt.ylabel('% Accuracy', fontsize=16)
plt.legend(['Train', 'Validation'], fontsize=16)
plt.show()
```

```python
plt.savefig(fname='TF_plots/Vary_Num_Layers.png', format='png')
```

## Variable Number of Nodes Data

#### Import all the data from the txt file and then part it as needed

```python
# Get all the traing and validation accuracy data
Nodes_tr = np.loadtxt(fname='TF_datasets/FFNN_VarNodes_Acc_data.txt')[:, [0, 2, 4, 6, 8, 10]]
Nodes_val = np.loadtxt(fname='TF_datasets/FFNN_VarNodes_Acc_data.txt')[:, [1, 3, 5, 7, 9, 11]]
```

#### Plot the Nodes Data to a fig

```python
nodes= 2**np.arange(6, 12)
fig = plt.figure()
plt.plot(nodes, Nodes_tr[-1, :]*100, 'k.--', nodes, Nodes_val[-1, :]*100, 'b.--', markersize=16)
plt.xlabel('Num Nodes', fontsize=16)
plt.ylabel('% Accuracy', fontsize=16)
plt.legend(['Train', 'Validation'], fontsize=16)
plt.show()
```

```python
plt.savefig(fname='TF_plots/Vary_Nodes.png', format='png')
```

## Variable Learning Rate

```python
# Get all the traing and validation accuracy data
Lrn_tr = np.loadtxt(fname='TF_datasets/FFNN_VarLearnRates_Acc_data.txt')[:, 2*np.arange(0,19)]
Lrn_val = np.loadtxt(fname='TF_datasets/FFNN_VarLearnRates_Acc_data.txt')[:, 2*np.arange(0,19)+1]
```

#### Plot the Learning Rate Data

```python
learning_rates = np.array([0.0001, 0.0003, 0.0005, 0.0007, 0.001, 0.003, 0.005, 0.007, 0.01, 0.013, 0.015, 0.017, 0.02, 0.025, 0.03, 0.07, 0.1, 0.2, 0.4])
fig = plt.figure()
plt.semilogx(learning_rates[:-4], Lrn_tr[-1, :-4]*100, 'k.--', learning_rates[:-4], Lrn_val[-1, :-4]*100, 'b.--', markersize=16)
plt.xlabel('Learning Rate', fontsize=16)
plt.ylabel('% Accuracy', fontsize=16)
plt.legend(['Train', 'Validation'], fontsize=16)
plt.show()
```

```python
plt.savefig(fname='TF_plots/Vary_LearningRates.png', format='png')
```

## Variable Batch Sizes

```python
# Get all the traing and validation accuracy data
batch_tr = np.loadtxt(fname='TF_datasets/FFNN_VarBatches_Acc_data.txt')[:, 2*np.arange(0,12)]
batch_val = np.loadtxt(fname='TF_datasets/FFNN_VarBatches_Acc_data.txt')[:, 2*np.arange(0,12)+1]
```

#### Plot the batch sizes data

```python
batches= 2**np.arange(0, 12)
fig = plt.figure()
plt.semilogx(batches, batch_tr[-1, :]*100, 'k.--', batches, batch_val[-1, :]*100, 'b.--', markersize=16)
plt.xlabel('Batch Size', fontsize=16)
plt.ylabel('% Accuracy', fontsize=16)
plt.legend(['Train', 'Validation'], fontsize=16)
plt.show()
```

```python
plt.savefig(fname='TF_plots/Vary_Batches.png', format='png')
```

## Save more Data to file

For ease of replotting these things at a later time and for allowing another person to add these plot to other plots of the same type, we will not just save only the data found in the plots to text files. The saved data will all have the same format in that the first column be the training accuracy data and the second column will be the validation accuracy data. 

```python
## epochs data
np.savetxt(fname='TF_datasets/Epochs_plt_data.txt', X=X_eps, header='Training Acc. - Validation Acc,')
# Variable number of layers data
layers_out = np.hstack((layer_tr[-1, :].reshape(-1,1),layer_val[-1, :].reshape(-1,1) ))
np.savetxt(fname='TF_datasets/Var_Layers_plt_data.txt', X=layers_out, header='Training Acc. - Validation Acc,')
# Variable number of nodes data
nodes_out = np.hstack((Nodes_tr[-1, :].reshape(-1,1),Nodes_val[-1, :].reshape(-1,1) ))
np.savetxt(fname='TF_datasets/Var_Nodes_plt_data.txt', X=nodes_out, header='Training Acc. - Validation Acc,')
# Variable learning Rate data
Lrn_out = np.hstack((Lrn_tr[-1, :].reshape(-1,1),Lrn_val[-1, :].reshape(-1,1) ))
np.savetxt(fname='TF_datasets/Var_LrnRate_plt_data.txt', X=Lrn_out, header='Training Acc. - Validation Acc,')
# Variable batch size data
batches_out = np.hstack((batch_tr[-1, :].reshape(-1,1),batch_val[-1, :].reshape(-1,1) ))
np.savetxt(fname='TF_datasets/Var_Batches_plt_data.txt', X=batches_out, header='Training Acc. - Validation Acc,')
```

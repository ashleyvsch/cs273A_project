 This jupyter notebook serves to combine the tensorflow and pytorch training and validation accuracy results determined while exploring the data set. The pytorch data was determined by Ashley and the tensorflow data was determined by Erik. We first combine the data coming for the 30 epochs test 

```python
# Normal Python imports
import numpy as np
import matplotlib.pyplot as plt
```

### 30 Epochs Plots and Fig
The line of code use to import the pytorch data is the code Ashley wrote in the PytorchFashionMNIST.ipynb file to import the data, parse and then plot it. I (Erik) will make slight edits to it should I need to 

```python
# first import th pytorch data
# epoch data results so you don't have to run test again
with open('exploration_data.txt') as f:
    lines = f.readlines()
    for i, line in enumerate(lines):
        lines[i] = line.strip('\n').strip()

epochs = list(map(int, lines[1][7:].split(' ')))
train_accuracies = list(map(float, lines[2][6:].split(' ')))
valid_accuracies = list(map(float, lines[3][11:].split(' ')))
lines = None
# import the TF data now
tf_epochs = np.loadtxt(fname='TF_datasets/Epochs_plt_data.txt')
# Now plot the data 
plt.rcParams.update({'font.size': 18})
plt.plot(epochs, train_accuracies, 'ko-', label = 'Pytorch Train')
plt.plot(epochs, valid_accuracies, 'ko--', label = 'Pytorch Val.')
plt.plot(epochs, tf_epochs[:,0]*100, 'ro-', label='TF Train')
plt.plot(epochs, tf_epochs[:,1]*100, 'ro--', label='TF Val.')
plt.xlabel('Epochs'); plt.ylabel('% Accuracy')
# plt.legend(); 
plt.show()
```

### Varing Opt. Learning Rate

```python
# Import the pytorch data first
with open('exploration_data.txt') as f:
    lines = f.readlines()
    for i, line in enumerate(lines):
        lines[i] = line.strip('\n').strip()

learning_rates = list(map(float, lines[18][:].split(' ')[1:]))
train_accuracies = list(map(float, lines[19][:].split(' ')[1:]))
valid_accuracies = list(map(float, lines[20][:].split(' ')[1:]))
lines = None
# import the TF data now
tf_lrn = np.loadtxt(fname='TF_datasets/Var_LrnRate_plt_data.txt')
# Now plot the data 
plt.rcParams.update({'font.size': 18})
plt.semilogx(learning_rates[:-4], train_accuracies[:-4], 'ko-', label = 'Pytorch Train')
plt.semilogx(learning_rates[:-4], valid_accuracies[:-4], 'ko--', label = 'Pytorch Val.')
plt.plot(learning_rates[:-4], tf_lrn[:-4,0]*100, 'ro-', label='TF Train')
plt.plot(learning_rates[:-4], tf_lrn[:-4,1]*100, 'ro--', label='TF Val.')
plt.xlabel('Learning rate'); plt.ylabel('% Accuracy')
# plt.legend(); 
plt.show()
```

### Varing Opt. Batch Sizes

```python
# Import the pytorch data first
with open('exploration_data.txt') as f:
    lines = f.readlines()
    for i, line in enumerate(lines):
        lines[i] = line.strip('\n').strip()

batch_sizes = list(map(float, lines[13][:].split(' ')[1:]))
train_accuracies = list(map(float, lines[14][:].split(' ')[1:]))
valid_accuracies = list(map(float, lines[15][:].split(' ')[1:]))
lines = None
# import the TF data now
tf_batches = np.loadtxt(fname='TF_datasets/Var_Batches_plt_data.txt')
# Now plot the data 
fig, ax1 = plt.subplots()

color = 'k'
ax1.set_xlabel('Batch Size')
ax1.set_ylabel('% Accuracy', color=color)
ax1.semilogx(batch_sizes, train_accuracies, 'ko-', label = 'Pytorch Train')
ax1.semilogx(batch_sizes, valid_accuracies, 'ko--',  label = 'Pytorch Val.')
ax1.semilogx(batch_sizes,  tf_batches[:,0]*100, 'ro-', label='TF Train')
ax1.semilogx(batch_sizes, tf_batches[:,1]*100, 'ro--', label='TF Val.')
ax1.tick_params(axis='y', labelcolor=color)
# ax1.legend(loc=2)

x = batch_sizes[6]
y = valid_accuracies[7]
label = "BS = {}".format(x)

plt.annotate(label, (x,y), textcoords="offset points", xytext=(-19,25), ha='center') 
```
